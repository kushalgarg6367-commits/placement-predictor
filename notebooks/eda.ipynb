```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-whitegrid') if 'seaborn-v0_8-whitegrid' in plt.style.available else None

df = pd.read_csv('../data/placement_data.csv')
df.head()
```

```python
df.info()
df.describe()
```

### Your insight:
_(e.g. any surprising ranges, missing values, class balance you notice from `.info()`/`.describe()`)_

```python
placement_rate = df['placed'].mean()
print(f"Overall placement rate: {placement_rate:.1%}")
df['placed'].value_counts().plot(kind='bar', title='Placed vs Not Placed (counts)')
plt.xticks([0, 1], ['Not Placed', 'Placed'], rotation=0)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df.boxplot(column='cgpa', by='placed', ax=axes[0])
axes[0].set_title('CGPA by placement outcome')
axes[0].set_xticklabels(['Not Placed', 'Placed'])

df.boxplot(column='coding_score', by='placed', ax=axes[1])
axes[1].set_title('Coding score by placement outcome')
axes[1].set_xticklabels(['Not Placed', 'Placed'])

plt.suptitle('')
plt.tight_layout()
plt.show()
```

### Your insight:
_(Which of these two looks like a stronger separator between placed/not placed? Any overlap that surprises you?)_

```python
branch_rates = df.groupby('branch')['placed'].mean().sort_values(ascending=False)
branch_rates.plot(kind='bar', title='Placement rate by branch')
plt.ylabel('Placement rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
branch_rates
```

### Your insight:
_(Does this match what you'd expect from real placement trends at MNIT or similar colleges? Where does it differ?)_

```python
backlog_rates = df.groupby('backlogs')['placed'].mean()
backlog_rates.plot(kind='bar', title='Placement rate by number of backlogs', color='indianred')
plt.ylabel('Placement rate')
plt.xticks(rotation=0)
plt.show()
```

```python
numeric_cols = ['cgpa', 'internships', 'projects_completed', 'backlogs',
                 'coding_score', 'communication_score', 'certifications',
                 'extracurricular', 'placed']
corr = df[numeric_cols].corr()

fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
ax.set_xticks(range(len(numeric_cols))); ax.set_xticklabels(numeric_cols, rotation=45, ha='right')
ax.set_yticks(range(len(numeric_cols))); ax.set_yticklabels(numeric_cols)
plt.colorbar(im)
ax.set_title('Correlation heatmap')
plt.tight_layout()
plt.show()

corr['placed'].sort_values(ascending=False)
```
